# HSV COLOR Space , Balloons

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# read the image
image=cv2.imread('images/water_balloons.jpg')

# make a copy of that image
image_copy=np.copy(image)

# Change color to RGB from BGR
image=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

## Plot color channels


```python
# RGB channels
r=image[:,:,0]
g=image[:,:,1]
b=image[:,:,2]

f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('BLue')
ax3.imshow(b,cmap='gray')
```

```python
## Convert from RGB to HSV
hsv=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

#HSV channels
h=hsv[:,:,0]
s=hsv[:,:,1]
v=hsv[:,:,2]
f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Hue')
ax3.imshow(h,cmap='gray')

```

## Define pink and hue selection threshold

```python
# Define our color selection criteria in HSV value

lower_hue=np.array([160,0,0])
upper_hue=np.array([180,255,255])
```

```python
#define our color selection criteria in RGB values

lower_pink=np.array([180,0,100])
upper_pink=np.array([255,255,230])
```

## mask the image


```python
# define the masked area  in RGB space
mask_rgb=cv2.inRange(image,lower_pink,upper_pink)

# mask the image
masked_image=np.copy(image)
masked_image[mask_rgb==0]=[0,0,0]

# visualize the mask
plt.imshow(masked_image)
```

```python
# Now try HSV!

# Define the masked area in HSV space
mask_hsv=cv2.inRange(hsv,lower_hue,upper_hue)

# mask the image
masked_image=np.copy(image)
masked_image[mask_hsv==0]=[0,0,0]

# visualizr the mask
plt.imshow(masked_image)
```
